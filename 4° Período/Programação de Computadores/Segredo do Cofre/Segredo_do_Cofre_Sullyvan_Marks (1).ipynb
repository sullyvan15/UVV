```
import numpy
import numpy as np
import random
```

```
def CriarPartida():
  global senha
  global partida
  partida = numpy.zeros((14,4), dtype=int)
  senha = np.random.permutation(random.sample(range(0, 9), 4))
 
```

```
def Partida():
  global senha
  global partida
  global nome
  global contadorPonto
  

  contador = 0
  contadorPonto = 10
  LinhaDica = -1
  LinhaPalpite = -2
  nome = input("Qual é o seu nome?")
  

    

  while contador < 7:

    dica = list()
    palpite = int(input("\nInsira 4 números, sem repetição:"))
    print('_________________________________________________\n')

      # Divisor de cada número do palpite para 
    p1 = palpite // 1000 % 10
    p2 = palpite // 100 % 10
    p3 = palpite // 10 %10
    p4 = palpite // 1 % 10
    


      # Verifica se os valores de palpite estão diferentes
    if p1 != p2 and p1 != p3 and p1 != p4:
      if p2 != p3 and p2 != p4:
        if p3 != p4:
                
      # comparação de valores de senha e palpite para adicionar as dicas
          if p1 in senha and p1 != senha[0]:
            dica1 = -1
            dica.append(dica1)

          elif p1 in senha and p1 == senha[0]:
            dica1 = 1
            dica.append(dica1)

          elif p1 not in senha:
            dica1 = 0
            dica.append(dica1)

          if p2 in senha and p2 != senha[1]:
            dica2 = -1
            dica.append(dica2)

          elif p2 in senha and p2 == senha[1]:
            dica2 = 1
            dica.append(dica2)

          elif p2 not in senha:
            dica2 = 0
            dica.append(dica2)

          if p3 in senha and p3 != senha[2]:
            dica3 = -1
            dica.append(dica3)

          elif p3 in senha and p3 == senha[2]:
            dica3 = 1
            dica.append(dica3)

          elif p3 not in senha:
            dica3 = 0
            dica.append(dica3)

          if p4 in senha and p4 != senha[3]:
            dica4 = -1
            dica.append(dica4)

          elif p4 in senha and p4 == senha[3]:
            dica4 = 1
            dica.append(dica4)

          elif p4 not in senha:
            dica4 = 0
            dica.append(dica4)


      # Contadores       
        contador += 1 
        contadorPonto -= 1
        LinhaDica +=2
        LinhaPalpite +=2

      # Adiciona o palpite e a dica na partida
      partida[LinhaDica, :] =  dica
      partida[LinhaPalpite, 0] = p1
      partida[LinhaPalpite, 1] = p2  
      partida[LinhaPalpite, 2] = p3  
      partida[LinhaPalpite, 3] = p4
      
      # print do palpite e da dica
      print(f'Palpite número {contador}:')
      print(f'{p1} {p2} {p3} {p4}')         
      print(f'{dica[0]} {dica[1]} {dica[2]} {dica[3]}')

        # Condições de acerto ou erro
      if p1 == senha[0] and p2 == senha[1] and p3 == senha[2] and p4 == senha[3]:
        print(f'Parabéns, você acertou!\n Sua pontuação: {contadorPonto}')
        
        break
      if contador == 7:
        if p1 != senha[0] or p2 != senha[1] or p3 != senha[2] or p4 != senha[3]:
          contadorPonto = 0
          print(f'Fim de Jogo!\n A senha era: {senha}\n Sua pontuação: {contadorPonto}')
        break
```
